# MEMBUAT HASIL BINER UNTUK ANALISIS FITUR

```python
data['final_result_2'] = data['final_result'].apply(lambda x: 'Failed' if \
    (x == 'Withdrawn' or x == 'Fail') else 'Succeeded')
```

```python
data['gender'].unique()
```

```python
gender = data[['id_student', 'gender']].drop_duplicates().groupby('gender').\
    agg({'id_student':'count'}).reset_index().\
    rename(columns={"gender": "Gender", 'id_student': 'Quantity'})

gender['Percentual'] = gender.apply(lambda x: round(100 * (x['Quantity'] / gender['Quantity'].sum()),2), axis=1)

fig = px.bar(gender, x='Gender', y='Quantity', \
    text='Quantity', color='Gender',  hover_data=['Quantity'],\
        template="seaborn")

fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Distribusi - Genre',
    xaxis_title="Genre",
    yaxis_title="Percentual [%]",
    legend_title="Genre",
    width=600,
    height=400,
    uniformtext_minsize=10,
    uniformtext_mode='hide',)
fig.show()
```

# Hubungan Antara Gender dan Hasil Akhir

```python
df = data.groupby(['gender', 'final_result']).agg({'id_student':'count'}).reset_index().\
    rename(columns={"gender": "Gender", "final_result": 'Result','id_student': 'Quantity'})

df['Percentual'] = df.apply(lambda x:  round(100 * (x['Quantity'] / df[df['Result'] == x['Result']]['Quantity'].sum()),2), axis=1)

fig = px.bar(df, x="Result", y="Percentual", text="Percentual", color="Gender",
            hover_data=['Percentual'], barmode = 'stack', template="seaborn")

fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Hasil berdasarkan Gender',
    xaxis_title="Result",
    yaxis_title="Percentual [%]",
    legend_title="Gender",
    width=600,
    height=400,
    uniformtext_minsize=10,
    uniformtext_mode='hide',
)
fig.show()
```

```python
# Hasil Biner - Lulus dan Gagal

df = data.groupby(['gender', 'final_result_2']).agg({'id_student':'count'}).reset_index().\
    rename(columns={"gender": "Gender", "final_result_2": 'Result','id_student': 'Quantity'})

df['Percentual'] = df.apply(lambda x:  round(100 * (x['Quantity'] / df[df['Result'] == x['Result']]['Quantity'].sum()),2), axis=1)

fig = px.bar(df, x="Result", y="Percentual", text="Percentual", color="Gender",
            hover_data=['Percentual'], barmode = 'stack', template="seaborn")

fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Hasil berdasarkan Gender',
    xaxis_title="Result",
    yaxis_title="Percentual [%]",
    legend_title="Gender",
    width=600,
    height=400,
    uniformtext_minsize=10,
    uniformtext_mode='hide',
)
fig.show()
```

# Berdasarkan Usia

```python
len(data['age_band'].unique())
```

```python
age_band = data[['age_band', 'id_student']].drop_duplicates().groupby('age_band').agg({'id_student':'count'}).reset_index().\
    rename(columns={"age_band": "age_band", 'id_student': 'Quantity'}).sort_values(by='Quantity', ascending=False)

age_band['Percentual'] = age_band.apply(lambda x: round(100 * (x['Quantity'] / age_band['Quantity'].sum()), 2) ,axis=1)

fig = px.bar(age_band, x='age_band', y='Quantity', text='Percentual', color ='age_band',  template="seaborn")
fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Distribusi - Usia',
    xaxis_title="age_band",
    yaxis_title="Quantity [siswa]",
    showlegend=False,
    width=800,
    height=550,
    uniformtext_minsize=10,
    uniformtext_mode='hide',
)

fig.show()
```

```python
df = data.groupby(['age_band', 'final_result']).agg({'id_student':'count'}).reset_index().\
    rename(columns={"age_band": "age_band", "final_result": 'Result','id_student': 'Quantity'})

df['Percentual'] = df.apply(lambda x:  round(100 * (x['Quantity'] / df[df['Result'] == x['Result']]['Quantity'].sum()),2), axis=1)

fig = px.bar(df, x="Result", y="Percentual", text="Percentual", color="age_band",
            hover_data=['Percentual'], barmode = 'stack', template='seaborn')
fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Hasil berdasarkan Usia',
    xaxis_title="Result",
    yaxis_title="Percentual [%]",
    legend_title="age_band",
    width=800,
    height=700,
    uniformtext_minsize=9,
    uniformtext_mode='hide',
)

fig.show()
```
